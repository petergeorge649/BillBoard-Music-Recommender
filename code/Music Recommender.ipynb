```python
 #1. importing libraries # BeautifulSoup/ Beautiful Soup is a Python package for parsing (analyse (a string or text) into logical syntactic components) HTML documents
from bs4 import BeautifulSoup
import requests
import pandas as pd


# 2. find url and store it in a data frame
url = "https://www.billboard.com/charts/hot-100"

# 3. download html with a get request
response = requests.get(url)
response.status_code # means has successfully passed

```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
lst = soup.select("span.chart-element__information > span.chart-element__information__artist.text--truncate.color--secondary")
lst
```

```python
lst_artists = soup.select("span.chart-element__information > span.chart-element__information__artist.text--truncate.color--secondary")
for artist in lst_artists: 
    print(artist.text)
```

```python
song = []
artist = []
```

```python
soup 
```

```python
songs_hot=soup.select("span.chart-element__information")
```

```python
len_songs=len(songs_hot)
len_songs
```

```python
for i in range(len_songs):
     song.append(soup.select("span.chart-element__information__song")[i].text)
     artist.append(soup.select("span.chart-element__information__artist")[i].text)
   
```

```python
song
```

```python
artist
```

```python
songs_100 = pd.DataFrame({'song':song,'artist':artist})
songs_100.head(100)
```

```python
def recommender():
    from random import randint
    #input from user
    song=input("Pick a song from the playlist ")

    if len(song)==0 or song == "0":
        print("Lost in thoughts?")
    else: 
        song=song.lower() #convert input to lowercase
        songs_100["song"]=songs_100["song"].apply(lambda x:x.lower())
        filter_song=songs_100[songs_100["song"]==song]
        if len(filter_song) ==0:
            print("Peter,Ibrahim,Ismail and Bright says the songs can not be found on my play list")

        else:
            print("That's a hot song")
            random_song = randint(0, len(songs_100)-1)
            print("You might also like " + songs_100["song"][random_song] + " by " + songs_100["artist"][random_song])
            # %s string modifier
            billboard_position = random_song + 1
            print("Which is number %s  on Hot_100 billboard" %billboard_position)
```

```python
recommender()
```
